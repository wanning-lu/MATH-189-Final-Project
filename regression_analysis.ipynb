```python
# create new DataFrames merging every two rows
#
# before_covid    Contains merged data from 2016-2019, before COVID
# after_covid     Contains merged data from 2020-2023, after COVID
# all             Contains merged data across all 8 years

before_covid = pd.DataFrame(columns=['City', 'Income', 'GPA', 'All', 'African American', 'Hispanic/ Latinx', 'Asian', 'White', 'Female', 'Male'])
after_covid = pd.DataFrame(columns=['City', 'Income', 'GPA', 'All', 'African American', 'Hispanic/ Latinx', 'Asian', 'White', 'Female', 'Male'])
all = pd.DataFrame(columns=['City', 'Income', 'GPA', 'All', 'African American', 'Hispanic/ Latinx', 'Asian', 'White', 'Female', 'Male'])

before_years = ['2016', '2017', '2018', '2019']
after_years = ['2020', '2021', '2022', '2023']
categories = ['All', 'African American', 'Hispanic/ Latinx', 'Asian', 'White', 'Female', 'Male']

for idx in range(0, 277, 2):
    before_values = {}
    after_values = {}
    all_values = {}
    
    # city
    before_values['City'] = after_values['City'] = all_values['City'] = merged_years['2016'].iloc[idx]['City']

    # admitted proportions
    before_adm = before_app = after_adm = after_app = all_adm = all_app = 0

    for category in categories:
        for year in before_years:
            before_adm += merged_years[year].iloc[idx][category]
            before_app += merged_years[year].iloc[idx+1][category]
            all_adm += merged_years[year].iloc[idx][category]
            all_app += merged_years[year].iloc[idx+1][category]
            
        for year in after_years:
            after_adm += merged_years[year].iloc[idx][category]
            after_app += merged_years[year].iloc[idx+1][category]
            all_adm += merged_years[year].iloc[idx][category]
            all_app += merged_years[year].iloc[idx+1][category]

        if before_app == 0:
            before_app = 1
        if after_app == 0:
            after_app = 1
        if all_app == 0:
            all_app = 1
    
        before_values[category] = before_adm / before_app
        after_values[category] = after_adm / after_app
        all_values[category] = all_adm / all_app

    # weighted gpa
    before_total_adm = after_total_adm = before_gpa = after_gpa = all_total_adm = all_gpa = 0
    
    for year in before_years:
        before_total_adm += merged_years[year].iloc[idx]['All']
        all_total_adm += merged_years[year].iloc[idx]['All']
        before_gpa += merged_years[year].iloc[idx]['Measure Values'] * merged_years[year].iloc[idx]['All']
        all_gpa += merged_years[year].iloc[idx]['Measure Values'] * merged_years[year].iloc[idx]['All']
        
    for year in after_years:
        after_total_adm += merged_years[year].iloc[idx]['All']
        all_total_adm += merged_years[year].iloc[idx]['All']
        after_gpa += merged_years[year].iloc[idx]['Measure Values'] * merged_years[year].iloc[idx]['All']
        all_gpa += merged_years[year].iloc[idx]['Measure Values'] * merged_years[year].iloc[idx]['All']

    before_values['GPA'] = before_gpa / before_total_adm
    after_values['GPA'] = after_gpa / after_total_adm
    all_values['GPA'] = all_gpa / all_total_adm

    # average income over the years
    before_inc = after_inc = all_inc = 0
    
    for year in before_years:
        before_inc += merged_years[year].iloc[idx]['Household Income']
        all_inc += merged_years[year].iloc[idx]['Household Income']
        
    for year in after_years:
        after_inc += merged_years[year].iloc[idx]['Household Income']
        all_inc += merged_years[year].iloc[idx]['Household Income']

    before_values['Income'] = before_inc / 4
    after_values['Income'] = after_inc / 4
    all_values['Income'] = all_inc / 8

    before_concat = pd.DataFrame(before_values, index=[0])
    after_concat = pd.DataFrame(after_values, index=[0])
    all_concat = pd.DataFrame(all_values, index=[0])

    before_covid = pd.concat([before_covid, before_concat])
    after_covid = pd.concat([after_covid, after_concat])
    all = pd.concat([all, all_concat])
```

## Data across all years, aggregated

```python
import statsmodels.api as sm 

X = all.drop(['All', 'City'], axis=1)
y = all['All']

model = sm.Logit(y, X)

result = model.fit()
```
